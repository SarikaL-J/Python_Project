```python
#importing Libraries
import pandas as pd  #import all libraries
import matplotlib.pyplot as plt 
import ast
import seaborn as sns
#loading dataset from csv file 
df = pd.read_csv("/Users/jastisarika/Downloads/data_jobs.csv")
#data Cleanup 
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

Filter down for US Data Analyst Roles

```python

df_da_usa = df[(df["job_title_short"]=='Data Analyst') & (df["job_country"]=='United States')]
df_da_usa #from here we ahve to pick up few columns in order to perform intial eda 
```

```python
#lets see the top 10 job locations 
df_da_usa["job_location"].value_counts().head(10)
#convert to daframe 
df_plot = df_da_usa["job_location"].value_counts().head(10).to_frame()
df_plot
```

```python
#plot a bar graph to visualise this data 
sns.set_theme(style = "ticks")
sns.barplot(data=df_plot,x="count",y="job_location",hue="count",palette='dark:b_r',legend=False)
sns.despine()  #to remove the border box 
plt.title("Number of Jobs for Data Analyst in USA")
plt.xlabel("Number of Jobs")
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1 , 3)
dict_column ={
    "job_work_from_home" :'Work from home',
    'job_no_degree_mention' : 'Job degree req',
    'job_health_insurance' : 'Health insurance Offered'
}

fig.set_size_inches((12,5))
for i,(column,title) in enumerate(dict_column.items()):         
          ax[i].pie(df_da_usa[column].value_counts(),startangle = 90,autopct = '%1.1f%%',labels =[False,True])   
          ax[i].set_title(title)

plt.suptitle("Benefit Analysis of Data Analyst Jobs",fontsize =16)
plt.show()
```

Companies to Explore:

```python
df_plot = df_da_usa["company_name"].value_counts().head(10).to_frame()
sns.set_theme(style = "ticks")
sns.barplot(data=df_plot,x="count",y="company_name",hue="count",palette='dark:b_r',legend=False)
sns.despine()  #to remove the border box 
plt.title("Number of Companies for Data Analyst in USA")
plt.xlabel("Number of Jobs")
plt.ylabel('')
plt.show()
```
