What are the most demanded skills for the top 3 most popular roles 

Methodology

1.Clean up skill column 

2.calculate skill count based on job_title_short

3.calculate skill percentage 

4.plot final findings


```python
#import all libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import ast
import seaborn as sns 
df = pd.read_csv("/Users/jastisarika/Downloads/data_jobs.csv")  #load the csv file
#Data Cleanup
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
#we want to do the analysis specific for united states

df_us = df[df['job_location']=='United States']
df_us
```

```python
# #now use explode method to breaddwon those job skills and assign it to a new dataframe df_skills
# df_skills = df_us.explode("job_skills")
# #now lets visualise how for each job_title , skills are brokendown
# df_skills[["job_title_short",'job_skills']]

df_skills = df_us.explode("job_skills")
skills_count = df_skills.groupby(['job_title_short','job_skills']).size()
df_skills_count = skills_count.reset_index(name="skill_count")
df_skills_count.sort_values(by="skill_count",ascending=False,inplace=True)
df_skills_count

```

```python
#how to get the top 3 most popular job titles
job_titles =df_skills_count["job_title_short"].unique()  #This is give us all the unique job_titles in an array 
job_titles = df_skills_count["job_title_short"].unique().tolist() #converts the array in to list 
job_titles = job_titles[:3]  #get the top 3 
job_titles =sorted(job_titles[:3])
job_titles
```

```python
#next step is plotting, we need 3 seperate plotting for the length of the above list 

fig, ax = plt.subplots(len(job_titles),1)
for i , job_title in enumerate(job_titles):
     df_plot = df_skills_count[df_skills_count["job_title_short"]==job_title].head(5)   #define a datframe we wan to plot 
     df_plot.plot(kind='barh',x='job_skills',y='skill_count',ax =ax[i],title = job_title)
     ax[i].invert_yaxis()
     ax[i].set_ylabel('')
     ax[i].legend().set_visible(False)
fig.suptitle('Counts of top skills in top postings',fontsize=15)
fig.tight_layout()
plt.show()
```

```python
# #now we need to cinverst this skill in to %. Perform agg 
# % of skills in job posting = skill count in job posting / total count of job posting 
df_us["job_title_short"].value_counts()
df_job_title_count = df_us["job_title_short"].value_counts().reset_index(name ='Jobs_total') #find total count of jobs 
```

```python
#now we need to merge this df in to df_skills_count on job_titel-short
df_skill_perc = pd.merge(df_skills_count , df_job_title_count,how = 'left',on ='job_title_short')
df_skill_perc   #now we want to find out skill percentage 
df_skill_perc["skill Percent"] = df_skill_perc["skill_count"]/df_skill_perc["Jobs_total"] *100
df_skill_perc
```

```python
fig, ax = plt.subplots(len(job_titles),1)
sns.set_theme(style='ticks')

for i , job_title in enumerate(job_titles):
     df_plot = df_skill_perc[df_skill_perc["job_title_short"]==job_title].head(5)   #define a datframe we wan to plot 
     # df_plot.plot(kind='barh',x='job_skills',y='skill Percent',ax =ax[i],title = job_title)
     sns.barplot(data=df_plot,x='skill Percent',y='job_skills',ax=ax[i],hue ='skill_count',palette='dark:b_r')
     # ax[i].invert_yaxis()
     ax[i].set_title(job_title)
     ax[i].set_ylabel('')
     ax[i].set_xlabel('')
     ax[i].legend().set_visible(False)
     ax[i].set_xlim(0,90)
     # sns.despine() 
     for n , v in enumerate(df_plot["skill Percent"]):                 #index n and value v 
          ax[i].text(v + 1, n, f'{v:.0f}%', va ='center')              #we need to specify x value to plot over and y value
     if i != len(job_titles) -1:                  #we need to remove the x axis ticks for forst 2 graphs and keep only for the last one 
         ax[i].set_xticks([])

fig.suptitle('Likelihood of skills requested in top postings',fontsize=15)
fig.tight_layout(h_pad =0.5)
plt.show()
```
