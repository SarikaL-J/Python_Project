```python
df["job_posted_date"]=pd.to_datetime(df.job_posted_date)
```

```python
df["job_posted_month"] = df.job_posted_date.dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts().sort_index()
```

```python
monthly_counts
```

```python
type(monthly_counts)
```

```python
plt.plot(monthly_counts.index,monthly_counts)
plt.show()   #line charts are used to plot xaxis- continous value(date or time) against count or number
```

```python
#now we have to calc how many job titles are posted.. 
df.info()
```

```python
job_counts = df["job_title_short"].value_counts()
```

```python
job_counts =job_counts.sort_values(ascending = False)
```

```python
plt.bar(job_counts.index,job_counts)
plt.title("Postings by Job Title")
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
plt.barh(job_counts.index,job_counts)
plt.title("Postings by Job Title")
plt.show()
```

```python
#now we have to calculate median of all slaries grouoped by the job titles 
df.info()
```

```python
grouped_values = df.groupby('job_title_short')["salary_year_avg"].median().sort_values(ascending = True)
```

```python
plt.barh(grouped_values.index, grouped_values)
plt.title("Median Salaries by Title")
```
