#copy -make a copy this objects indices and data
a new object will be created with a copy of calling object's data and indices. Any modifications to the data or indices  of the copy will not be reflected int he original data . lets understand by using example

```python
df_altered = df
```

```python
df_altered.loc[0:5,'salary_year_avg']
```

```python
median_salary_year = df_altered['salary_year_avg'].median()
median_salary_year
```

```python
df['salary_year_avg'] = df['salary_year_avg'].fillna(median_salary_year)
```

```python
df_altered.loc[0:5,'salary_year_avg']
```

```python
df  # see here df is also altered or modified as df_altered..because id(df) = id(df_altered) becos, df=df_altered
#  that is why we have to create a copy of datafram so that original and copy will be diff
```

```python
id(df)
```

```python
id(df_altered)  #both are same 
```

```python
df_altered = df.copy()
```
