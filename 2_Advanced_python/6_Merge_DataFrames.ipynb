load a csv file containing other job postings info 

```python
df_USA_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
```

```python
%store -r df_usa_pivoted
#as we have created pivot table in the previus exceroce we can use this her 
```

```python
df_USA_software_pivot
```

```python
df_merged_jobs = df_usa_pivoted.merge(df_USA_software_pivot, on='job_posted_month')
df_merged_jobs
```

```python
top_5 = df_merged_jobs.sum().sort_values(ascending=False).head(5).index.to_list()
```

```python
top_5
```

```python
df_merged_jobs[top_5].plot(kind='line')
plt.title('Monthly Job Postings for the Tech Jobs in the USA')
plt.xlabel('2022')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
